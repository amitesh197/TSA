```python
import pandas as pd
from pandasql import sqldf
import numpy as  np
```

```python
from pandasql import sqldf

Query_string = """ select * from dfpsql limit 5 """
sqldf(Query_string, globals())
```

```python
Query_string = """ select * from dfpsql order by SmartCity_Index Desc;"""
sqldf(Query_string, globals())
```

```python
Query_string = """ select count(Smart_People) as count, 
min(Smart_People) as min, max(Smart_People) as max, 
avg(Smart_People) as mean from dfpsql;"""
sqldf(Query_string, globals())
```

```python
Query_string = """ select ID , City, Country, sum(SmartCity_Index) as Sum_SmartCity_Index from dfpsql group by ID;"""
sqldf(Query_string, globals())
```

```python
Query_string = """ select City , avg(SmartCity_Index) as mean, min(SmartCity_Index) as min, max(SmartCity_Index) as max from dfpsql group by ID;"""
sqldf(Query_string, globals())
```

```python
data2 = {'ID': [1011, 1017, 1013], 'Country': ['Canada', 'Denmark', 'Norway'], 'SmartCity_Index_relative_Edmonton': [1000, -1000, -2500]}
Dept_df = pd.DataFrame(data2, columns = ['ID', 'Country', 'SmartCity_Index_relative_Edmonton'])
Dept_df
```

```python
Query_string = """ select * from dfpsql a INNER JOIN Dept_df b ON a.Country = b.Country;"""
sqldf(Query_string, globals())
```

```python
Query_string = """ select * from dfpsql a LEFT JOIN Dept_df b ON a.Country = b.Country;"""
sqldf(Query_string, globals())
```

```python
Query_string = """ select * from dfpsql a left OUTER JOIN Dept_df b ON a.Country = b.Country;"""
sqldf(Query_string, globals())
```
